# Generate $\mathcal{H}^2$ matrix with [h2tools](https://bitbucket.org/muxas/h2tools) package

```python
from __future__ import absolute_import
```

```python
from __future__ import print_function, absolute_import, division
import numpy as np
import scipy.sparse as sps
from h2tools.collections import triangular_surface as trisurf
from h2tools.problem import Problem, ClusterTree
from h2tools.mcbh import mcbh
import imp
h2ts = imp.load_source('h2ts', '../h2_to_sparse.py')

block_size = 25
tau = 1e-3
iters = 1
onfly = 0
verbose = 0
func = trisurf.integral_inverse_r3


def load_mesh(file_name, verbose=1):
    f = open(file_name, 'r')
    n_v = int(f.readline().strip())
    print ("Number of vert.", n_v)
    v = np.zeros((3, n_v), dtype=float)
    for i in range(n_v):
        v[:, i] = f.readline().strip().split(",")
    n_t = int(f.readline().strip())
    print ("Number of tr.", n_t)
    t = np.zeros((3, n_t), dtype=int)
    for i in range(n_t):
        t[:, i] = f.readline().strip().split(",")
    return v, t
v, t = load_mesh('mesh_16.txt')
data = trisurf.TriangularSurface(v, t)
tree = ClusterTree(data, block_size)
problem = Problem(func, tree, tree, symmetric=1, verbose=verbose)
matrix = mcbh(problem, tau=tau, iters=iters, onfly=onfly, verbose=verbose)
```

# Compress $\mathcal{H}^2$ matrix and orthogonalize transition matrices

```python
matrix.svdcompress(tau=1e-5)
```

# Compute the non-extensive sparse factorization of the $\mathcal{H}^2$ matrix.

```python
S, U, V = h2ts.convert_h2_to_sparse(matrix, check_error=1)
```

# Show the sparse matrix

```python
import matplotlib as mpl
import matplotlib.pyplot as plt
%matplotlib inline
plt.figure(figsize=(10, 10))
plt.spy(S.res, markersize=0.1)
```
